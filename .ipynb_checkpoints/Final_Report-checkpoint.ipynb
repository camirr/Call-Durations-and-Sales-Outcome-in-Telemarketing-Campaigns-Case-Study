### 1.1 Our Question:
A previous study has shown that the last contact duration that needed to be rescheduled to obtain a final answer from the client is one of the most relevant attributes that determine the outcome of the telemarketing campaign ("success" or "failure") (Moro, Sérgio et al., 2014). Moreover, as will be illustrated in the boxplots in section 2.1.2, there is an apparent difference in the median last call duration between successful and failed outcomes in our dataset. Hence, we are going to compare whether there exists a significant difference in **last contact duration** between a successful outcome and a failed outcome among those customers with university degrees. 

Data on the last call duration were stored in the "duration" column in the dataset, and customers' education levels were recorded in the "education" column. Outcomes of telemarketing campaigns were stored in the "y" column of the dataset. 

As will be illustrated in section 2.1.4, the sample distributions of the call durations are right skewed for both successful and failed outcomes, and there exist many outliers that might negatively affect our result. To cope with the uncertainty of outliers, for this report, we will use both **difference in medians** and **difference in means** of the last contact duration between successful and failed outcomes as location parameters. And we will use the standard deviation and IQR as scale parameters. 

### 1.2 Our Dataset:

We will use the Bank Marketing Data Set retrieved from the UCI Learning Respotory. This dataset is related to telemarketing campaigns of a Portuguese banking institution, where the campaigns were based on phone calls. Each record contains several numerical variables (e.g., education, contact duration, three-month Euribor) and some categorical variables (e.g., y, which indicates contact outcome ("success" or "failure")). We will filter out customers with a university degree and only focus on the "duration" and "y" columns.

## 2.2 Methods

Now, let's test our expected preliminary findings: the call duration of the successful session is longer than the call duration of the failed session. Thus, we are only concerned about the one-sided hypothesis test.

As the sample size of our data is large enough, both bootstrapping and the asymptotic method are suitable for finding the specific numerical patterns and obtaining a convincing result. We will perform tests with **both** methods in this report and conduct confidence intervals to estimate the precision of our parameter value.

Since we want to minimize the potential error caused by the presence of outliers, we will first use the bootstrapping method to examine the difference in the median of the two results. Then, we will use the asymptotic method (two-sample t-test) to examine the difference in the means of the two results after eliminating the outliers. (The reason we use the t-test for the mean here instead of the median is because it is covered in this course and we have more confidence in its correctness.) Finally, we will compare and discuss two methods' results.

### 2.2.1 Method: Bootstrapping

To compare the distribution, we select the **median** as the variable of interest. With such method, we further conduct our analysis as follow.

#### Hypothesis testing:

We want to compare the difference in medians of call durations between successful and failure outcomes. In order to achieve our goal, we want to test the null hypothesis H0 against alternative hypothesis H1. We first start by defining the two sample medians:

- m0: median call duration of successful sales

- m1: median call duration of failed sales

Then we define the null and alternative hypothesis as follow:

- H0:  m0 - m1 = 0 
(the median call duration of successful sales is same as the median call duration of failed sales)

- H1: m0 - m1 > 0 
(the median call duration of successful sales is greater than the median call duration of failed sales)

We will use a 90% confidence level (for a right tail test at the significant level of 5%) for our hypothesis testing. 

#### A) Test Statictics

First, the observed median difference between two groups in terms of call duration is computed.

```r
sample_medians_diff <- sample_estimates%>%
                    select(y, median)%>%
                    pivot_wider(names_from = y, values_from = median)%>%
                    transmute(diff = yes - no)%>%
                    pull(diff)

sprintf("The difference in medians of between call duration of a successful outcome and a failure is: %.2f", sample_medians_diff )
```

#### B) Bootstrap sampling distribution

Second, the sample median within each bootstrapped sample is derived to generate a bootstrap sampling distribution.

```r
set.seed(1000)

null_model_bank <- bank_data_sample %>%
                specify(formula = duration ~ y) %>% 
                hypothesize(null = "independence") %>% 
                generate(reps = 1000, type = "permute") %>% 
                calculate(stat="diff in medians", order = c("yes", "no"))

head(null_model_bank)
```


Table 2.2.1: First 6 rows of bootstrap sampling distribution

#### C) P-value

Third, by plugging the observed statistics into the model, the p-value can be calculated with a right tail test at the significant level of 0.05. 

Visualize the bootstrap sampling distribution of difference of medians of duration with shaded p-value.

```r
options(repr.plot.width = 15, repr.plot.height = 8)
bank_result_plot <- null_model_bank %>%
                    visualize() + 
                    shade_p_value(obs_stat = sample_medians_diff, direction = "right")+
                    theme(text = element_text(size = 20))+
                    xlab("Difference in median of duration")
bank_result_plot
```

Figure 2.2.1: Bootstrap sampling distribution of difference of medians with shaded p-value

```r
p_value_duration = null_model_bank %>%
                    get_p_value(obs_stat = sample_medians_diff, direction = "right")
```

The **p-value** of difference in medians is:

```r
pull(p_value_duration)
```

Since we have generated 1000 samples, we will record the p value to be 0.001. Since 0.001 < 0.50, the null hypothesis is rejected and our test is statistically significant (the differences between two groups are significant) as the p value is 0.001 and the shaded areas of plot is really small.

#### D) Confidence Intervals

Our hypothesis test has confirmed that our findings are significant, we now find the corresponding 90% conidence interval to interpret our findings.

```r
set.seed(1111)

bootstrap_distribution <- bank_data_sample %>%
                specify(formula = duration ~ y) %>% 
                generate(reps = 1000, type = "bootstrap") %>% 
                calculate(stat="diff in medians", order = c("yes", "no"))

duration_ci_0.9 <- bootstrap_distribution %>%
                    get_confidence_interval(type = 'percentile', level = 0.9)
duration_ci_0.9 
```

Table 2.2.2: Upper and lower bounds of 90% CI

#### E) Visualization of the result

Finally, with the numerical features detected previously, we visualize the result and determine whether we reject the null hypothesis or not. 

```r
options(repr.plot.width = 15, repr.plot.height = 8)
shade_ci_duration <- bootstrap_distribution%>%
                    visualize(bins = 20)+
                    shade_confidence_interval(endpoints = duration_ci_0.9)+
                    theme(text = element_text(size = 20))+
                    labs(x = "Difference in median of duration")+
                    ggtitle("Percentile-based 90% confidence interval")
shade_ci_duration
```

Figure 2.2.2: Bootstrap sampling distribution of difference of medians with shaded 90% CI

Therefore, we are 90% confident that the interval [216, 251] contains the population difference of medians of duration of successful and failed sessions.

### 2.2.2 Method: Asymptotics

As we mentioned in the preliminary result section, our two samples were heavily right skewed and contained many outliers. The means would be heavily influenced by outliers, which makes their values less useful measures of the average. However, we still want to examine the mean of the sample. After we have eliminated the outliers, we determine the confidence intervals of the mean differences using asymptotics and compare the confidence intervals of these two different measures.

To compare the distribution, we select the **mean** as the variable of interest. With such method, we further conduct our analysis as follow.

### Hypothesis testing:

We want to compare the difference in means of call durations between successful and failure outcomes. We first start by defining the two sample means:

- m0: mean call duration of successful sales

- m1: mean call duration of failed sales

Then we define the null and alternative hypothesis as follow:

- H0:  m0 - m1 = 0 
(the mean call duration of successful sales is same as the median call duration of failed sales)

- H1: m0 - m1 > 0 
(the mean call duration of successful sales is greater than the median call duration of failed sales)

Again, we will use a 90% confidence level (for a right tail test at the significant level of 5%) for our hypothesis testing. 

We fisrt delete the outliers present in both the yes and no dataset. Then, compute sample estimates

```r
# YES
# Q1 = 232
# Q3 = 678
# IQR = 446
# upper outlier =  Q3 + 1.5IQR
# lower outlier =  Q1 - 1.5IQR

yes_sample_no_outliers <- yes_sample %>%
                filter(duration < (678 + 1.5*446) & duration > (232 - 1.5*446))

yes_no_outliers <- yes_sample_no_outliers%>%
                summarize(mean = mean(duration),
                          sd = sd(duration),
                          n = n())

# NO
# Q1 = 92
# Q3 = 264
# IQR = 172
# upper outlier =  Q3 + 1.5IQR
# lower outlier =  Q1 - 1.5IQR
no_sample_no_outliers <-no_sample %>%
                filter(duration < (264 + 1.5*172) & duration > (92 - 1.5*172))

no_no_outliers <- no_sample_no_outliers %>%
                summarize(mean = mean(duration),
                          sd = sd(duration),
                          n = n())
```

Now, let's use asymptotic CLT.

First, we start by checking conditions of our sample. Although the pattern of normal distribution cannot be found from the sample distribution, we can apply the central limit theroem. Our dataset satisfies the condition of applying CLT, as the sample size is large enough (greater than 10000 which is sufficient).

#### A) Test Statictics

Second, compute the observed two-sample t-statistic.

```r
yes_mean <- yes_no_outliers$mean
yes_sd <- yes_no_outliers$sd
yes_n <- yes_no_outliers$n

no_mean <- no_no_outliers$mean
no_sd <- no_no_outliers$sd
no_n <- no_no_outliers$n



obs_two_sample_t <- (yes_mean - no_mean)/sqrt(yes_sd^2/yes_n + no_sd^2/no_n)

sprintf("The difference in means of between call duration of a successful outcome and a failure is: %.2f", obs_two_sample_t)
```

#### B) P -Value

Third, compute and visualize the p-value.

```r
df <- (yes_sd^2/yes_n + no_sd^2/no_n)^2 / ((yes_sd^4/(yes_n^2*(yes_n - 1))) + (no_sd^2/(no_n^2*(no_n -1))))   
    
p_value_t <- pt(obs_two_sample_t, df, lower.tail = FALSE)
```

The **p-value** of difference in means is:

```r
p_value_t
```

The p value obtained with a two-sample t-test is also approximately to 0, which has the same as the bootstrap method. 

Use t.test() to verify the result (p-value)

```r
t_test_result <- tidy(t.test(x = yes_sample_no_outliers$duration,
                             y = no_sample_no_outliers$duration,
                             alternative = 'greater'))
t_test_result
```

Table 2.2.3: Summary of t-test

#### C) Confidence Interval

Finally, obtain the 90% confidence interval and visualize the result.

```r
yes_var <-(yes_sd/sqrt(yes_n))^2
no_var <-(no_sd/sqrt(no_n))^2

bank_clt_ci <- 
    tibble(lower_ci = yes_mean - no_mean - qnorm(0.95) *sqrt(yes_var + no_var),
           uppper_ci = yes_mean - no_mean + qnorm(0.95) * sqrt(yes_var + no_var))
bank_clt_ci
```

Table 2.2.4: Upper and lower bounds of 90% CI using CLT

Therefore, we are 90% confident that the interval [274.836, 299.1891] contains the population difference of means of duration of successful and failed sessions.

### 3.1 Summary of the result

**Comparison between the two methods:**

The p-values calculated by both methods are less than **0.05.** Thus, we conclude that there is sufficient evidence to reject **H0** for both difference in means and medians, assuming that both difference in means and medians call durations of successful outcomes are **greater than** the  both difference in means and medians of call durations of failed outcomes for clients with a university degree.

This suggests that a type I error is possible.

90% confidence interval for difference in medians with bootstrapping: **(216,251)**

90% confidence interval for the difference of the means using the asymptotic method (outliers eliminated): **(274.836, 299.1891)**

The differences in p-values and confidence intervals are due to the following reasons:
1. The bootstrapping method examines the difference in medians, while the asymptotic method examines the difference in means.
2. The use of different samples. The asymptotic method eliminated the outliers, while the bootstrap method kept the sample as it was.

We believe that the asymptotics method is more appropriate and trustworthy than the bootstrapping method for this situation. Despite the fact that both methods gave the same result, namely that H0 was rejected, the bootstrapping method simply gave 0 for the p-value, which is technically incorrect since the type I error will never be zero. The asymptotic method, on the other hand, gave us an exact p-value that is very close to zero and yet more accurate than the bootstrapping method.

**Successful phone calls of longer duration are a natural phenomenon, and this is exactly what we expect.**
People prefer to get more information on the phone when they want to subscribe to something. 
And the customers who don't want to subscribe in the future will try to end the call as soon as possible.
