### 1.0 Background:
Marketing sales campaign is a typical strategy used by companies to grow their business. In particular, telemarketing conducted remotely has enabled marketing to optimize customer lifetime value by evaluating available information and metrics, hence allowing the company to build longer and tighter relationships in alignment with business demand.

### 1.1 Our Question:
Previous study has shown that the last contact duration that needed to be rescheduled to obtain a final answer from the client, is the one of the most relevant attributes that determine the outcome of the telemarketing campaign ("success" or "failure") (Moro, Sérgio, et al., 2014). Moreover, as will be illustrated in the boxplots in section 2.1.2, there exists an appearent difference in median of last call duration between successful and failed outcomes in our existing dataset. 

Hence, we are going to comparing whether there exist significant differences in last contact duration between a successful outcome and a failed outcome among those customors with University degrees. Data on the last call duration were stored in the “duration” column in the dataset, and customors' education levels were recorded in the "education" column. Outcomes of telemarketing campaigns were stored in the "y" of column of the dataset. 

As will be illustrated in section 2.1.4, the sample distributions of the call durations of are right skewed for both successful and failed outcomes. Hence, for this report, we will use the difference of median of the last contact duration between successful outcomes and failed outcomes as the location parameter and the IQR as the scale parameter.

### 1.2 Our Dataset:

We will use the Bank Marketing Data Set retrived from the UCI Learning Respotory. This dataset is related to telemarketing campaigns of a Portuguese banking institution, where the campaigns were based on phone calls. Each record contains several numerical variables (e.g. education, contact duration, three-month Eutibor) and some categorical variables (e.g. y which indicates contact outcome ("success" or "failure")). We will filter out custermer with a university degree and only focus on the "duration" and "y" column.

The sample distribution of duration of last call and the preliminary results above show that there is a difference in duration of last call between successful and failed campaign results. However, they are not sufficient enough to determine whether the difference in median is significant. Therefore, a hypothesis test for differences between the medians of successful and failed outcomes is conducted. Confidence intervals will be constructed to determine a range of plausible differences.  IQR will be calculated to indicate the shape and spread of the distribution of the difference in the median. 

## 2.2 Methods

According to the preliminary result, both bootstrapping and the asymptotic method are appropriate for this study. The sample size of our data is large enough to find the specific numerical patterns and get a convincing result.

Thus, we will conduct tests using **both** methods in the report and compare their results.

#### A) Hypothesis testing:

Now,let's test our expected preliminary findings, the call duration of successful session is longer than the call duration of failed session, thus we are only concerned about the one-sided hypothesis test. 

We want to compare the difference in median of call duration between successful and failure outcomes. In order to achieve our goal, we want to test the null hypothesis H0 against alternative hypothesis H1. We first start by defining the two sample medians:

- m0: median call duration of successful sessions

- m1: median call duration of failed sessions

Then we define the null and alternative hypothesis as follow:

- H0:  m0 - m1 = 0 
(the median call duration of successful sessions is same as the median call duration of failed sessions)

- H1: m0 - m1 > 0 
(the median call duration of successful sessions is greater than the median call duration of failed sessions)

We will use a 90% confidence interval (for a right tail test at the significant level of 0.05) for our hypothesis testing. 

### 2.2.1 Method: Bootstrap

To compare the distribution, we select the **median** as the variable of interest. With such method, we further conduct our analysis as follow.

#### A) Test Statictics

First, the observed median difference between two groups in terms of call duration is computed.

```r
# Compute difference in sample medians between call duration of a successful outcome and a failure
sample_medians_diff <- sample_estimates%>%
                    select(y, median)%>%
                    pivot_wider(names_from = y, values_from = median)%>%
                    transmute(diff = yes - no)%>%
                    pull(diff)
```

The **difference in medians** of between call duration of a successful outcome and a failure is:

```r
sample_medians_diff
```

#### B) Bootstrap sampling distribution

Second, the sample median within each bootstrapped sample is derived to generate a bootstrap sampling distribution.

```r
# Generate the bootstrap sampling distribution
set.seed(1000)

null_model_bank <- bank_data_sample %>%
                specify(formula = duration ~ y) %>% 
                hypothesize(null = "independence") %>% 
                generate(reps = 1000, type = "permute") %>% 
                calculate(stat="diff in medians", order = c("yes", "no"))

head(null_model_bank)
```


Table 2.2.1: First 6 rows of bootstrap sampling distribution

#### C) P-value

Third, by plugging the observed statistics into the model, the p-value can be calculated with a right tail test at the significant level of 0.05. 

```r
# Visualize the bootstrap sampling distribution of difference of medians of duration with shaded p-value 
options(repr.plot.width = 15, repr.plot.height = 8)
bank_result_plot <- null_model_bank %>%
                    visualize() + 
                    shade_p_value(obs_stat = sample_medians_diff, direction = "right")+
                    theme(text = element_text(size = 20))+
                    xlab("Difference in median of duration")
bank_result_plot
```

Figure 2.2.1: Bootstrap sampling distribution of difference of medians with shaded p-value

```r
# Compute p value
p_value_duration = null_model_bank %>%
                    get_p_value(obs_stat = sample_medians_diff, direction = "right")
```

The **p-value** of difference in medians is:

```r
pull(p_value_duration)
```

The null hypothesis is rejected and our test is statistically significant ( the differences between two groups are significant) as the p value is 0 and the shaded areas of plot is really small.

#### D) Confidence Interval

We also find the corresponding 90% conidence interval to interpret our findings in different ways.

```r
set.seed(1111)

bootstrap_distribution <- bank_data_sample %>%
                specify(formula = duration ~ y) %>% 
                hypothesize(null = "independence") %>% 
                generate(reps = 1000, type = "bootstrap") %>% 
                calculate(stat="diff in medians", order = c("yes", "no"))

# Compute the corresponding 90% confidence interval 
duration_ci_0.9 <- bootstrap_distribution %>%
                    get_confidence_interval(level = 0.9)
duration_ci_0.9 
```

Table 2.2.2: Upper and lower bounds of 90% CI

#### E) Visualization of the result

Finally, with the numerical features detected previously, we visualize the result and determine whether we reject the null hypothesis or not. 

```r
# Visualize bootstrap sampling distribution of difference of medians of duration with shaded 90% confidence interval
options(repr.plot.width = 15, repr.plot.height = 8)
shade_ci_duration <- bootstrap_distribution%>%
                    visualize(bins = 20)+
                    shade_confidence_interval(endpoints = duration_ci_0.9)+
                    labs(x = "Difference in median of duration")+
                    ggtitle("Percentile-based 90% confidence interval")
shade_ci_duration
```

Figure 2.2.2: Bootstrap sampling distribution of difference of medians with shaded 90% CI

Therefore, we are 90% confident that the interval [-7.525, 7] contains the population difference of medians of duration of successful and failed sessions.

### 2.2.2 Method: Asymptotics

To compare the distribution, we select the mean as the variable of interest. With such method, we further conduct our analysis as follow.

Now, let's use asymptotic CLT.

###First Check conditions: Although the pattern of normal distribution cannot be found from the sample distribution, we can apply the CLT. Our dataset satisfy the condition of applying CLT as the sample size is large enough.

#### A) Test Statictics

Second, compute the observed two-sample t-statistic.

```r
obs_two_sample_t <- bank_data_sample %>%
                    specify(formula = duration ~ y) %>% 
                    calculate(stat="t", order = c("yes", "no"))%>%
                    as.numeric()
obs_two_sample_t
```

#### B) P -Value

Third, compute and visualize the p-value.

```r
yes_mean <- sample_estimates$mean[2]
yes_sd <- sample_estimates$sd[2]
yes_n <- sample_estimates$n[2]

no_mean <- sample_estimates$mean[1]
no_sd <- sample_estimates$sd[1]
no_n <- sample_estimates$n[1]

df <- (yes_sd^2/yes_n + no_sd^2/no_n)^2 / ((yes_sd^4/(yes_n^2*(yes_n - 1))) + (no_sd^2/(no_n^2*(no_n -1))))   
    
p_value_t <- pt(obs_two_sample_t, df, lower.tail = FALSE)
p_value_t
```

The p value obtained with a two-sample t-test is also approximately to 0, which meets the bootstrap sampling distribution. 

Use t.test() to check the result (p-value)

```r
yes_sample <- 
    bank_data_sample %>% 
    filter(y == 'yes') %>% 
    pull(duration)

no_sample <- 
    bank_data_sample %>% 
    filter(y == 'no') %>% 
    pull(duration)

t_test_result <- tidy(t.test(x = yes_sample,
                             y = no_sample,
                             alternative = 'greater'))
t_test_result
```

Table 2.2.3: Summary of t-test

#### C) Confidence Interval

Finally, obtain the 90% confidence interval and visualize the result.

```r
yes_var <- (yes_sd/sqrt(yes_n))^2
no_var <- (no_sd/sqrt(no_n))^2

#Obtain the 90% confidence interval for the difference in means of yes and no call duration using CLT.
bank_clt_ci <- 
    tibble(lower_ci = yes_mean - no_mean - qnorm(0.95) *sqrt(yes_var + no_var),
           uppper_ci = yes_mean - no_mean + qnorm(0.95) * sqrt(yes_var + no_var))
bank_clt_ci
```

Table 2.2.4: Upper and lower bounds of 90% CI using CLT
